## Eleman Temelli


```python
def naive_add(x,y):
  assert len(x.shape)==2#x ve y 2 boyutlud
  assert x.shape==y.shape
  x=x.copy()
  for i in range(x.shape[1]):
    for j in range(x.shape[1]):
      x[i,j]+=y[i,j]
  return x    
```

```python
x=np.array([[2,2,2],
            [7,14,21],
            [8,16,24]])
y=np.array([[2,2,2],
            [7,14,21],
            [8,16,24]])
naive_add(x,y)3,
```

## Yayma Operasyonu


```python
def naive_add_matrix_vector(x,y):
  assert len(x.shape)==2
  assert len(y.shape)==1
  assert x.shape[1]==y.shape[0]
  x=x.copy()
  for i in range(x.shape[0]):
    for j in range(x.shape[1]):
      x[i,j]+=y[j]
  return x
```

```python
x=np.array([[7,14,21],
            [7,14,21],
            [7,14,21]]
           )
y=np.array([1,2,3])
z=naive_add_matrix_vector(x,y)
z
 
 
```
